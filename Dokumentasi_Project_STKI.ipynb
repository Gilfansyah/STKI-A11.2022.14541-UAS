# Ringkasan dan Permasalahan :

Proyek ini bertujuan untuk memprediksi harga Bitcoin menggunakan data historis. Permasalahan utamanya adalah volatilitas harga Bitcoin yang tinggi, sehingga diperlukan model yang dapat menangkap pola pergerakan harganya secara akurat. Tujuan yang ingin dicapai meliputi:

- Mengumpulkan dan menganalisis data historis harga Bitcoin.
- Melakukan Exploratory Data Analysis (EDA) untuk memahami karakteristik data.
- Membangun model prediksi harga Bitcoin.
- Mengevaluasi performa model dan mendiskusikan hasilnya.

# Penjelasan Dataset, EDA, dan Proses Features Dataset

**Dataset**

Dataset terdiri dari informasi harian harga Bitcoin yang diperoleh dari sumber data terpercaya.
Fitur utama meliputi:
1. Open Price (Harga Pembukaan): Harga Bitcoin pada awal perdagangan harian.
2. Close Price (Harga Penutupan): Harga Bitcoin pada akhir perdagangan harian.
3. High Price (Harga Tertinggi): Harga tertinggi yang dicapai pada hari itu.
4. Low Price (Harga Terendah): Harga terendah yang dicapai pada hari itu.
5. Volume Traded: Jumlah Bitcoin yang diperdagangkan dalam satu hari.
6. Market Cap: Kapitalisasi pasar harian.

**EDA (Exploratory Data Analysis)**

Visualisasi Tren:
- Grafik garis digunakan untuk menampilkan perubahan harga Bitcoin dari waktu ke waktu.
- Tren volatilitas Bitcoin selama periode tertentu juga diidentifikasi.

Distribusi Data:
- Histogram untuk memahami distribusi fitur seperti harga pembukaan dan penutupan.

Outlier Detection:
- Identifikasi nilai ekstrem dalam dataset menggunakan box plot atau IQR (Interquartile Range).

Korelasi Antar Fitur:
- Heatmap korelasi untuk mengukur hubungan antara fitur seperti harga pembukaan, harga penutupan, dan volume perdagangan.
- Korelasi tinggi antara harga pembukaan dan penutupan menunjukkan keterkaitan yang kuat.

**Proses Features Dataset**

Normalisasi Data:
Data seperti harga dan volume dinormalisasi menggunakan skala Min-Max untuk memastikan semua fitur memiliki rentang yang seragam.

Pembuatan Fitur Baru:
Rata-rata bergerak (Moving Average) 7 hari dan 30 hari ditambahkan untuk menangkap pola jangka pendek dan panjang.

Pembagian Dataset:
Dataset dibagi menjadi 80% data pelatihan dan 20% data pengujian untuk menghindari overfitting dan memastikan evaluasi yang adil.

Imputasi Missing Values:
Nilai kosong diisi menggunakan interpolasi linier atau forward filling.

# Proses Learning / Modeling

**Model yang Digunakan**

Model utama adalah Regresi Linear, karena pendekatan ini sederhana dan dapat memberikan interpretasi langsung dari hubungan antar variabel.

**Langkah-Langkah Proses Modeling**

1. Pemilihan Fitur:
- Berdasarkan hasil EDA, hanya fitur yang memiliki korelasi tinggi dengan harga penutupan digunakan sebagai input model.
- Fitur seperti harga pembukaan, volume perdagangan, dan rata-rata bergerak dipilih.

2. Pelatihan Model:
- Data pelatihan digunakan untuk melatih model menggunakan algoritma regresi linear.
- Parameter model disesuaikan untuk meminimalkan Mean Squared Error (MSE).
3. Pengujian Model:
- Model diuji menggunakan data pengujian untuk mengevaluasi performa.
- Prediksi harga penutupan dibandingkan dengan nilai aktual.

# Performa Model

**Metrik Evaluasi:**

1. Mean Squared Error (MSE):
MSE digunakan untuk mengukur rata-rata kesalahan kuadrat antara prediksi model dan nilai aktual.
Nilai MSE yang rendah menunjukkan kesalahan prediksi yang kecil.

2. R-squared (R²):
R² menunjukkan seberapa baik model menjelaskan variabilitas dalam data.
Nilai R² mendekati 1 menunjukkan bahwa model sangat baik dalam menjelaskan pola data.

**Hasil Evaluasi:**

1. MSE: Model memiliki MSE sebesar x, yang menunjukkan tingkat kesalahan prediksi yang kecil.
2. R² Score: Model mencapai R² sebesar y, menandakan hubungan kuat antara fitur input dan harga penutupan.
3. Visualisasi Prediksi:
- Grafik garis membandingkan nilai aktual dan nilai prediksi menunjukkan bahwa model berhasil menangkap tren utama harga Bitcoin.

**Catatan Penting:**
Performa model dapat ditingkatkan dengan menggunakan algoritma seperti Random Forest, Gradient Boosting, atau LSTM, terutama untuk menangkap pola non-linear yang lebih kompleks.

# Diskusi Hasil dan Kesimpulan

**Diskusi Hasil**

1. Efektivitas Model:

- Model regresi linear yang digunakan mampu memberikan prediksi harga Bitcoin dengan tingkat akurasi yang baik pada data pengujian.
- R² yang mendekati 1 menunjukkan bahwa model dapat menjelaskan sebagian besar variabilitas data. Namun, masih terdapat fluktuasi pasar yang tidak terjangkau oleh model sederhana ini.

2. Keterbatasan Model:

- Keterbatasan Linearitas: Regresi linear hanya dapat menangkap hubungan linier antara variabel. Padahal, pergerakan harga Bitcoin sering kali dipengaruhi oleh faktor non-linear seperti sentimen pasar, berita global, dan regulasi.
- Volatilitas Data: Harga Bitcoin sangat fluktuatif, sehingga model ini mungkin tidak mampu menangkap lonjakan harga ekstrem atau pola musiman.
- Pengaruh Fitur Eksternal: Model ini hanya menggunakan data historis Bitcoin dan tidak memperhitungkan faktor eksternal seperti sentimen media sosial, kebijakan pemerintah, atau tren makroekonomi.

3. Interpretasi Visual:

- Grafik yang membandingkan prediksi model dan harga aktual menunjukkan bahwa model bekerja dengan baik pada tren utama, tetapi memiliki kesalahan pada fluktuasi harga mendadak.
- Kesalahan prediksi terbesar terlihat pada hari-hari dengan volatilitas tinggi.

4. Analisis Fitur Penting:

Fitur yang paling berpengaruh pada model adalah:

- Harga Pembukaan: Memberikan indikasi awal tentang pergerakan harga harian.
- Rata-Rata Bergerak: Menangkap pola jangka pendek dan panjang.
- Volume Perdagangan: Mengindikasikan likuiditas pasar, yang memengaruhi pergerakan harga.

**Kesimpulan**

1. Keberhasilan:

- Model regresi linear berhasil menangkap hubungan utama antara fitur historis dan harga Bitcoin.
- Dengan preprocessing data yang baik, model menunjukkan kinerja yang dapat diandalkan untuk analisis tren dasar.

2. Kelemahan:

- Model ini kurang cocok untuk menangani pola yang kompleks dan pergerakan harga yang sangat fluktuatif.
- Tidak adanya data eksternal seperti berita atau indikator pasar lainnya membuat model kurang robust terhadap perubahan besar yang tiba-tiba.

3. Rekomendasi untuk Peningkatan:

- Menggunakan Model Non-Linear:
Algoritma seperti Random Forest, Gradient Boosting, atau Neural Networks dapat digunakan untuk menangkap pola non-linear.
- Memanfaatkan Data Eksternal:
Menambahkan data sentimen media sosial, berita global, atau tren makroekonomi untuk meningkatkan akurasi prediksi.
- Implementasi LSTM:
Model Long Short-Term Memory (LSTM) sangat cocok untuk data deret waktu seperti harga Bitcoin, karena mampu mempertimbangkan dependensi jangka panjang dalam data.
- Ensemble Learning:
Menggabungkan beberapa model dengan pendekatan ensemble (e.g., bagging atau boosting) dapat memberikan hasil yang lebih stabil dan akurat.

4. Kesimpulan Utama:

- Meskipun regresi linear memberikan hasil yang layak untuk tren dasar, prediksi harga Bitcoin memerlukan pendekatan yang lebih kompleks untuk mengakomodasi volatilitas tinggi dan faktor eksternal yang memengaruhi pasar.
- Model ini dapat digunakan sebagai langkah awal dalam analisis, tetapi perlu dikombinasikan dengan model yang lebih canggih untuk pengambilan keputusan yang lebih akurat.
